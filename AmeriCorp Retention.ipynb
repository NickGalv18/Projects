```python
import pandas as pd
```

```python
df = pd.read_csv('Copy of AC.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(100,75))
sns.countplot(data=df, x='District', hue='Completed', palette='husl', ax=axs[0, 1])
sns.countplot(data=df, x='Program', hue='Completed', palette='husl', ax=axs[1, 0])
sns.countplot(data=df, x='Returning', hue='Completed', palette='husl', ax=axs[1, 1])


axs[0, 1].set_xlabel('District', fontsize=30)
axs[0, 1].set_title('AmeriCorps Retention', fontsize=40)
axs[1, 0].set_xlabel('Program', fontsize=30)
axs[1, 0].set_title('AmeriCorps Retention', fontsize=40)
axs[1, 1].set_xlabel('Returning', fontsize=30)
axs[1, 1].set_title('AmeriCorps Retention', fontsize=40)

for ax in axs.flat:
    ax.tick_params(axis='x', labelsize=10)

for ax in axs.flat:
    for p in ax.patches:
        if p.get_height() > 0:
            ax.text(p.get_x() + p.get_width()/2., p.get_height(), '{:.1f}%'.format(p.get_height()/len(df)*100), 
                    ha='center', va='bottom', fontsize=14)

plt.show()
```
